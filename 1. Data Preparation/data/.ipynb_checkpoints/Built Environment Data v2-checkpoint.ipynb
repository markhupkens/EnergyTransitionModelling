### Lookup & Match
Energy data from the klimaatmonitor dataset (df_energy) needs to be matched with the 'housing type' column in the BAG data (DF), so we look up what building type the house has and match it with average E,G consumption per buildingtype per municipality for each house (N=7.8M) in the dataset.

```python
'''Add column average expected energy consumption per house'''

# Lookup housing type and match with average demand data per building type

df.loc[df['Housing Type']=='C1', 'G_average_p_type'] = df['Gemiddeld gasgebruik vrijstaande woning [m3] [2017]']
df.loc[df['Housing Type']=='C2', 'G_average_p_type'] = df['Gemiddeld gasgebruik twee-onder-een-kap-woning [m3] [2017]']
df.loc[df['Housing Type']=='C3', 'G_average_p_type'] = df['Gemiddeld gasgebruik hoekwoning [m3] [2017]']
df.loc[df['Housing Type']=='C4', 'G_average_p_type'] = df['Gemiddeld gasgebruik tussenwoning [m3] [2017]']
df.loc[df['Housing Type']=='C5', 'G_average_p_type'] = df['Gemiddeld gasgebruik appartement [m3] [2017]']
df.loc[df['Housing Type']=='C6', 'G_average_p_type'] = df['Gemiddeld gasgebruik appartement [m3] [2017]']

# Elek
df.loc[df['Housing Type']=='C1', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik vrijstaande woning [kWh] [2017]']
df.loc[df['Housing Type']=='C2', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik twee-onder-een-kap-woning [kWh] [2017]']
df.loc[df['Housing Type']=='C3', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik hoekwoning [kWh] [2017]']
df.loc[df['Housing Type']=='C4', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik tussenwoning [kWh] [2017]']
df.loc[df['Housing Type']=='C5', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik appartement [kWh] [2017]']
df.loc[df['Housing Type']=='C6', 'E_average_p_type'] = df['Gemiddeld elektriciteitsgebruik appartement [kWh] [2017]']

# Add housing matrix to enable building type count at the end of the script
df.loc[df['Housing Type']=='C1', 'House (detached, c1) (# houses)'] = 1
df.loc[df['Housing Type']=='C2', 'Two under one roof (c2) (# houses)'] = 1
df.loc[df['Housing Type']=='C3', 'Cornerhouse (c3) (# houses)'] = 1
df.loc[df['Housing Type']=='C4', 'Row house (c4) (# houses)'] = 1
df.loc[df['Housing Type']=='C5', 'Appartment (c5) (single floor) (# houses)'] = 1
df.loc[df['Housing Type']=='C6', 'Appartment (c6) (multiple floors) (# houses)'] = 1
```

### Aggregation
Now we have done all data manipulations on house-level, we will aggregate the data to municipal level for model-setup.

```python
'''Aggregate data for insights''' #good one

df.G_average_p_type = pd.to_numeric(df.G_average_p_type,errors='coerce') # Drop nonnumeric data

# Group data by municipality and aggregate
df_muni = df.groupby('Municipality Name').agg({'House No':'count','House (detached, c1) (# houses)':'count',
                                               'Two under one roof (c2) (# houses)':'count',
                                               'Cornerhouse (c3) (# houses)':'count',
                                               'Row house (c4) (# houses)':'count',
                                               'Appartment (c5) (single floor) (# houses)':'count',
                                               'Appartment (c6) (multiple floors) (# houses)':'count',
                                               'G_average_p_type':'mean',
                                               'E_average_p_type':'mean','Construction Year':'mean'})                                            
df_muni['Total E (GWh)'] = (df_muni['House No'] * df_muni['E_average_p_type'])/1000000
df_muni['Total G (Miljoen m3)'] = (df_muni['House No'] * df_muni['G_average_p_type'])/1000000

df_muni = df_muni.round(2)
# df_muni['Municipality Name'] = df_muni.index
df_muni.head()
```

```python
'''Aggregate data for insights'''

df.G_average_p_type = pd.to_numeric(df.G_average_p_type,errors='coerce') # Drop nonnumeric data

# Group data by municipality and aggregate
df_muni = df.groupby('Municipality Name').agg({'House No':'count','G_average_p_type':'mean','E_average_p_type':'mean',
                                              'House (detached, c1) (# houses)':'count',
                                               'Two under one roof (c2) (# houses)':'count',
                                              'Cornerhouse (c3) (# houses)':'count','Row house (c4) (# houses)':'count',
                                              'Appartment (c5) (single floor) (# houses)':'count',
                                              'Appartment (c6) (multiple floors) (# houses)':'count'})
df_muni['Total E (GWh)'] = (df_muni['House No'] * df_muni['E_average_p_type'])/1000000
df_muni['Total G (Miljoen m3)'] = (df_muni['House No'] * df_muni['G_average_p_type'])/1000000

df_muni.round(2) #.sort_values(by=index).head()
```

### Merge data with Model-setup framework
Import modelsetup file and merge build-environment data with it on municipality name

```python
df_gem = pd.read_excel("D:/markhupkens/Dropbox/EnTransitionNL/0. Data/ModelSetUpEnergieNL02 (1).xlsx",sheet_name='gemeente')
df_buurt = pd.read_excel("D:/markhupkens/Dropbox/EnTransitionNL/0. Data/ModelSetUpEnergieNL02 (1).xlsx",sheet_name='buurt')
df_wijk = pd.read_excel("D:/markhupkens/Dropbox/EnTransitionNL/0. Data/ModelSetUpEnergieNL02 (1).xlsx",sheet_name='wijk')
df_mod = pd.read_excel("D:/markhupkens/Dropbox/EnTransitionNL/0. Data/ModelSetUpEnergieNL02 (1).xlsx",sheet_name='ModelSpecification')

# split entity string to match building data on municipality name
df_gem['Municipality Name'] = df_gem["Entities"].str.split(" G").str[0]
```

```python
# merge data frames
df_gem = df_gem.merge(df_muni, left_on='Municipality Name', right_on= df_muni.index)
df_gem = df_gem.drop(['Municipality Name'], axis=1)
```
